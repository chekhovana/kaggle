```python
from common import *
from xgboost import XGBRegressor
import os
```

```python
x_train, y_train, x_test = read_preprocessed_data(postfix='preprocessed')
print(x_train.shape, y_train.shape, x_test.shape)
```

```python
params = {'n_estimators': 3458, 'min_child_weight': 0, 'max_depth': 5, 
          'learning_rate': 0.007, 'lambda': 81, 'gamma': 90, 
          'colsample_bytree': 0.4, 'subsample': 0.2, 'alpha': 159, 
          'objective': 'reg:squarederror'}
xgb = XGBRegressor(**params)
print(get_cv_score(x_train, y_train, xgb))
```

```python
folder = 'data'
xgb.fit(x_train, y_train)
preds = xgb.predict(x_test)
template = pd.read_csv(os.path.join(folder, 'sample_submission.csv'))
output = pd.DataFrame({'Id': template['Id'], 'SalePrice': preds})
output.to_csv(os.path.join(folder, 'submission.csv'), index=False)
```
